```python
import re
class Account:
    def __init__(self,balance=0):
        with open('Blackjack_Output.txt') as f:
            lines = f.readlines()
        match = re.findall(r"Bank Roll:\s(.\d+)",str(lines))
        for value in [int(x) for x in match]:
            balance = balance + value
        self.balance = balance
        match_two = re.findall(r"Hands:\s(.\d+)",str(lines))
        num_hands = 0
        for value in [int(x) for x in match_two]:
            num_hands = num_hands + value
        self.num_hands = num_hands
        match_three = re.findall(r"Wins:\s(.\d+)",str(lines))
        num_wins = 0
        for value in [int(x) for x in match_three]:
            num_wins = num_wins + value
        self.num_wins = num_wins
        match_four = re.findall(r"Losses:\s(.\d+)",str(lines))
        num_losses = 0
        for value in [int(x) for x in match_four]:
            num_losses = num_losses + value
        self.num_losses = num_losses
        match_five = re.findall(r"Pushes:\s(.\d+)",str(lines))
        num_pushes = 0
        for value in [int(x) for x in match_five]:
            num_pushes = num_pushes + value
        self.num_pushes = num_pushes
        self.bet = int()

    def clearbet(self):
        self.bet = 0
        
    def __str__(self):
        return f'Player Money Balance: ${self.balance}\n Number of Hands Played: {self.num_hands}'
        
    def deposit(self,dep_amt):
        self.balance += dep_amt
        print(f'${dep_amt} deposited into account. Balance is ${self.balance}')
        
    def withdraw(self,wd_amt):
        if self.balance >= wd_amt and wd_amt > 0:
            self.balance -= wd_amt
            print(f'Bet Accepted. Balance is ${self.balance}')
            return True
        else:
            print('Funds Unavailable for bet!')
            return False
        
```

```python
import random

class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    def __str__(self):
        return self.rank + ' of ' + self.suit
    
    def show(self):
        print("{} of {}" .format(self.rank, self.suit))

class Deck:
    
    def __init__(self):
        # Note this only happens once upon creation of a new Deck
        self.all_cards = [] 
        for suit in suits:
            for rank in ranks:
                # This assumes the Card class has already been defined!
                self.all_cards.append(Card(suit,rank))
    
    def __str__(self):
        dec_comp = ''
        for x in self.all_cards:
            dec_comp = dec_comp + x.__str__() +"\n"
        return 'The deck is: \n' + dec_comp
    
    def shuffle(self):
        # Note this doesn't return anything
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        # Note we remove one card from the list of all_cards
        return self.all_cards.pop()    
```

```python
def check_win(dealer,player):
    if get_total(dealer) > 21:
        print('Computer busts!')
    elif get_total(dealer) > get_total(dealer):
        print(f'Computer Wins! Computer has {get_total(dealer)} vs Player 1 has {get_total(player)}')
        return 2
    elif get_total(dealer) < get_total(player):
        print(f'Player 1 Wins! Computer has {get_total(dealer)} vs Player 1 has {get_total(player)}')
        return 1
    elif get_total(dealer) == get_total(player):
        print(f'Push! Computer has {get_total(dealer)} vs Player 1 has {get_total(player)}')
        return 3
```

```python
def get_total(a):
    total = 0
    aces = 0
    for x in a:
        total = total + x.value
        if x.rank == 'Ace':
            aces = aces + 1
    while aces > 0 and total > 21:
        total = total - 10
        aces = aces - 1
    return total
```
